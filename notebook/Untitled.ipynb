```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument("--headless")       # define headless
driver = webdriver.Chrome("D:\ProgramData\Anaconda3\envs\spider\Scripts\chromedriver.exe",chrome_options=chrome_options)
# driver = webdriver.Chrome(executable_path = "D:\ProgramData\Anaconda3\envs\spider\Scripts\chromedriver.exe",chrome_options="--headless")
```

```python
driver.get("https://morvanzhou.github.io/")
driver.set_window_size(960,1039)
driver.find_element_by_css_selector("#recent-update > li:nth-child(3) .lazy-img").click()
```
